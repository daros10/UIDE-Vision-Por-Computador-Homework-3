## - Cajas Delimitadoras

El presente entregable tiene como objetivo comprender el proceso de creación de cajas delimitadoras (bounding boxes) en objetos dentro de una imagen, utilizando coordenadas en los ejes X e Y para definir su posición y tamaño.

## - Desarrollo

Inicialmente, procedemos a cargar nuestra imagen de referencia para poder observar de mejora manera las coordenadas dentro del plano.

```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('1-Draw_Bounding_Box/resources/animals.png')
print('Image shape:', img.shape)

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.style.use("dark_background")
plt.imshow(img_rgb)
plt.show()
```

Una vez que la imagen ha sido cargada, se crean las cajas delimitadoras a partir de coordenadas en los ejes X e Y para los tres sujetos de prueba, en este caso un gato, un pájaro y un perro.

```python
x1_cat, y1_cat = 0, 470
x2_cat, y2_cat = 950, 1495

x1_bird, y1_bird = 1015, 1002
x2_bird, y2_bird = 1400, 1495

x1_dog, y1_dog = 1400, 215
x2_dog, y2_dog = 2755, 1495
```

Con las coordenadas definidas procedemos a crear los labels para identificar cada caja delimitadora.

```python
cat_color = (255, 255, 0)
bird_color = (0, 255, 0)
dog_color = (255, 0, 0)

cv2.putText(img_rgb, "Cat", (x1_cat, y1_cat - 20),
            cv2.FONT_HERSHEY_SIMPLEX, 2, cat_color, 5)

cv2.putText(img_rgb, "Bird", (x1_bird, y1_bird - 20),
            cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 255, 0), 5)

cv2.putText(img_rgb, "Dog", (x1_dog, y1_dog - 20),
            cv2.FONT_HERSHEY_SIMPLEX, 2, dog_color, 5)
```

Finalmente, presentamos y mostramos los resultados, con nuestros sujetos delimitados por cajas de colores.

```python
img_box = cv2.rectangle(img_rgb, (x1_cat, y1_cat), (x2_cat, y2_cat), cat_color, 15)
img_box = cv2.rectangle(img_rgb, (x1_bird, y1_bird), (x2_bird, y2_bird), bird_color, 15)
img_box = cv2.rectangle(img_rgb, (x1_dog, y1_dog), (x2_dog, y2_dog), dog_color, 15)

plt.style.use("dark_background")
plt.imshow(img_box)
plt.axis('off')
plt.show()
```
